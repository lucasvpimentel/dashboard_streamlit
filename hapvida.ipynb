```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import csv
```

# Time Series

```python
def data_column(df):
    df['DATA'] = ''
    df['DATA'] = df.apply(lambda x: f"{int(x['DIA']):02d}/{int(x['MES']):02d}/{int(x['ANO'])}", axis=1)
    df['DATA'] = pd.to_datetime(df['DATA'], format='%d/%m/%Y')
    df['MES_ANO'] = df['DATA'].dt.strftime('%m/%Y')
    return df
```

```python
df_hapvida = pd.read_csv('files/RECLAMEAQUI_HAPVIDA.csv')
df_hapvida.head()
```

```python
data_column(df_hapvida)
```

```python
df_hapvida.groupby('MES_ANO').size().reset_index(name='TOTAL_RECLAMACOES').to_csv('csv_treated/hapvida/time_series.csv')
```

# Por Estados

```python
def get_estado(df):
    df['ESTADO'] = df['LOCAL'].apply(lambda x: x.split(' - ')[1])
    df_estado = df.groupby('ESTADO').size().reset_index(name='TOTAL_RECLAMACOES')

    df_estado['ESTADO'] = df_estado['ESTADO'].replace('--', 'Na')

    df_estado.sort_values(by='TOTAL_RECLAMACOES', ascending=False, inplace=True)
    return df_estado
```

```python
df_estado_hapvida = get_estado(df_hapvida)
df_estado_hapvida.to_csv('csv_treated/hapvida/by_state.csv')
```

# Por Status

```python
df_hapvida.groupby('STATUS').size().reset_index(name='STATUS_RECLAMACOES').to_csv('csv_treated/hapvida/status.csv')
```
