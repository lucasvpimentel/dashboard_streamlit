# Série temporal

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
def data_column(df):
    df['DATA'] = ''
    df['DATA'] = df.apply(lambda x: f"{int(x['DIA']):02d}/{int(x['MES']):02d}/{int(x['ANO'])}", axis=1)
    df['DATA'] = pd.to_datetime(df['DATA'], format='%d/%m/%Y')
    df['MES_ANO'] = df['DATA'].dt.strftime('%m/%Y')
    return df

def get_count(df):
    df_count = df.groupby('MES_ANO').size().reset_index(name='TOTAL_RECLAMACOES')
    df_count['MES'] = df_count['MES_ANO'].apply(lambda x: int(x.split('/')[0]))
    df_count.sort_values(by='MES_ANO', inplace=True)
    return df_count
```

```python
df_ibyte = pd.read_csv('files/RECLAMEAQUI_IBYTE.csv')
df_ibyte
```

```python
data_column(df_ibyte)
```

```python
time_series = get_count(df_ibyte)
path = 'csv_treated/ibyte/'
time_series.to_csv(path + 'time_series.csv')
time_series
```

# Estado

```python
def get_estado(df):
    df['ESTADO'] = df['LOCAL'].apply(lambda x: x.split(' - ')[1])
    df_estado = df.groupby('ESTADO').size().reset_index(name='TOTAL_RECLAMACOES')

    df_estado['ESTADO'] = df_estado['ESTADO'].replace('--', 'Na')

    df_estado.sort_values(by='TOTAL_RECLAMACOES', ascending=False, inplace=True)
    return df_estado
```

```python
df_estado_ibyte = get_estado(df_ibyte)
```

```python
df_estado_ibyte['ESTADO'] = df_estado_ibyte['ESTADO'].replace({'C':'CE',
                                   'P': 'PA',
                                   'SEM-INFO': 'Na',
                                   'naoconsta': 'Na'})
```

```python
df_estado_ibyte = df_estado_ibyte.groupby('ESTADO')['TOTAL_RECLAMACOES'].sum().reset_index(name='TOTAL_RECLAMACOES')
df_estado_ibyte.sort_values(by='TOTAL_RECLAMACOES', ascending=False, inplace=True)
df_estado_ibyte.to_csv(path + 'by_state.csv')
df_estado_ibyte
```
